#### Numpy

1. Create an 12X4 array (denoted by A) filled with integers randomly draw from a range between 10 to 60 with equal probability.

```python
import numpy as np

# Create an 12X4 array filled with integers randomly drawn from a range between 10 to 60 with equal probability.
A = np.random.randint(10, 61, size=(12, 4))

print(A)
```

2. Sort array A by its second column in descending order

```python
# Sort A by its second column in descending order
sort_indices = np.argsort(A[:, 1])[::-1]
A_sort = A[sort_indices]

print(A_sort)
```

3. Sort array A by its fifth row in ascending order

```python
# Sort A by its fifth row in ascending order
sort_indices = np.argsort(A[4, :])
A_sort = A[:, sort_indices]

print(A_sort)
```

4. Delete the third column of A and insert a new row

```python
# Delete the third column of A 
A_delete = np.delete(A, 2, axis=1)

# insert a row consisting of all 20s to the end of A
row = np.full((1, A.shape[1]), 20)
A_insert = np.vstack((A, row))

print(A_delete)
print(A_insert)
```

#### Pandas

1. Load data in automobile_data.csv in the "data" subfolder of this assignment to create a Pandas data frame. Print out the top and bottom five rows of the data.

```python
import pandas as pd

# load automobile_data.csv as data frame
df = pd.read_csv('data/automobile_data.csv')

# print top 5 rows
print(df.head())

# print bottom 5 rows
print(df.tail())
```

2. Find the company (companies) that produce the top two most expensive cars

```python
# find and print out the name of the company (companies) that produce the top two most expensive cars
df_sort = df.sort_values(by='price', ascending=False)
top_two = df_sort['company'].head(2)

print(top_two)
```

3. Calculate the average horsepower of cars for each company

```python
# calculate the average horsepower of cars for each company
avg = df.groupby('company')['horsepower'].mean()

print(avg)
```

4. Sort all cars by length in descending order and save the sorted data to a csv file with name of automobile_data_sorted.csv

```python
# sort by length
df_length = df.sort_values(by='length', ascending=False)

# save the sorted data to automobile_data_sorted.csv
df_length.to_csv('data/automobile_data_sorted.csv')
```

#### Matplotlib

1. Draw a line plot of the hyperbolic tangent function. Comparing this plot with the one you drew by hand in Part1. 

```python
%matplotlib inline
import matplotlib.pyplot as plt

# draw a line plot of the hyperbolic tangent function
x = np.linspace(-5, 5, 100)
y = np.tanh(x)
plt.plot(x, y, label='tanh(x)')
plt.title('Hyperbolic Tangent Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.legend()
plt.show()
```

2. Plot a bar chart to show the numbe of cars produced by each company in automobile_data.csv. Your figure should have a size of 8-by-3 in inches and looks like the figure in below. Hint: use rcParams in matplotlib.pyplot to set the figure size. 
![ncars_by_company](images/ncars-by-company.png)

```python
# plot a bar chart to show the numbe of cars produced by each company
num_cars = df.groupby('company')['company'].count()

plt.rcParams['figure.figsize'] = (8, 3)

plt.bar(num_cars.index, num_cars.values)
plt.title('Number of cars produced by each company')
plt.xlabel('company')
plt.ylabel('Number of cars')
plt.xticks(rotation=45)

plt.grid(True, linestyle='--', linewidth=1)
plt.show()
```

3. The company_sales_data.csv in "data" subfolder contains the sales data of a company in each month of last year. Columns 2-7 provides the number of units sold for individual products. Please calculate the total sold units in last year for each product and show the data using a Pie chart. In this chart, display the number of units sold for each product in percentage. The Pie chart should look like this. 
![total_sales](images/total-sales.png)

```python
# plot a pie chart of total sold units in percentage for each product
df = pd.read_csv('data/sales_data.csv')

product_columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
total = df[product_columns].sum()

product_labels = ['FaceCream', 'FaceWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
plt.pie(total, labels=product_labels, autopct='%1.1f%%')
plt.title('Sales data')

plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.25))
plt.show()
```
